```python
#Concept Based Questions
1)Multicollinearity is a phenomenon in regression analysis where two or more independent variables in a statistical model 
are highly correlated.It impacts the regression model in several ways:
    ->Unstable Coefficients
    ->Difficulty in Interpretation
    ->Loss of Statistical Significance
    ->Inflated Standard Errors
    ->Reduced Predictive Power.
Challenges of Multicollinearity:
----Difficulty in Identifying Important Predictors: It becomes challenging to identify which variables are most important for
    predicting the outcome.
----Unstable Model Coefficients: Small changes in data can result in large variations in coefficient estimates, making it 
    hard to rely on the model.
Techniques to Handle Multicollinearity:
----Correlation Analysis:
----VIF (Variance Inflation Factor)
----Feature Selection
----Ridge Regression
----Principal Component Analysis (PCA)
----Data Collection and Domain Knowledge


2)Outliers are data points that are very different from the others. They can mess up your predictions in regression models. 
  To deal with them, you can:

  a. **Look at Your Data:** Use plots to spot these odd points.

  b. **Z-Score:** Calculate how far each point is from the average, and if it's too far, consider it an outlier.

  c. **IQR Method:** Create a range using the middle 50% of your data and see if any points are outside that range.

  d. **Residuals:** For regression, look at how far your predictions are from real data. Large differences might mean outliers.

  e. **Transform Data:** Change your data using math to reduce the impact of outliers.

  f. **Capping:** Set a limit for extreme values.

  g. **Robust Models:** Use models that are less bothered by outliers.

  h. **Data Imputation:** Replace outliers with sensible values.

  i. **Different Models:** Try models that handle outliers better.

  j. **Remove Outliers:** In extreme cases, you can throw them away, but be careful because you might lose important information.

    
3)Overfitting in regression happens when the model learns the training data too well, like memorizing a textbook rather than 
  understanding the material. It's a problem because the model doesn't work well with new data.

  To fix this, we use regularization techniques like Ridge and Lasso:

- **Ridge** makes sure that the numbers (coefficients) in the model don't get too big. It's like saying,
    "Don't give too much importance to any one thing."

- **Lasso** not only does that but can also say, "If something isn't important, forget about it completely."

  These techniques help the model find the right balance between fitting the training data well and not getting too complex. 
  That way, the model can predict new data better and is easier to understand.


4)When regular regression doesn't work for data that has curves or bends, you can use **polynomial regression**. 
  It's like using curved lines instead of straight lines to fit the data. 

- Imagine data points on a graph. Instead of drawing a straight line, polynomial regression draws curves, 
  allowing the model to understand wavy patterns.

- To pick the right curve, you choose a **degree**, which is how bendy the curve can be. Too little bend, and it doesn't 
  fit well. Too much bend, and it fits too well but only for the data it's seen before. You find the best degree using testing 
  and checking how well the curve works for new data.
```

```python
#task 1 (B)

import pandas as pd

# Read the data
df1 = pd.read_csv(r"C:\Users\aryan\Downloads\train.csv\train.csv")

# Filter the data for COVID-19 deaths
deaths_data = df1[df1['Target'] == 'Fatalities']

# Group the data by country/region and sum the death counts
country_deaths = deaths_data.groupby('Country_Region')['TargetValue'].sum()

# Find the country/region with the highest number of deaths
country_with_highest_deaths = country_deaths.idxmax()
highest_death_count = country_deaths.max()

print(f"The country or region with the highest number of COVID-19 deaths is {country_with_highest_deaths} with {highest_death_count} deaths.")
```
